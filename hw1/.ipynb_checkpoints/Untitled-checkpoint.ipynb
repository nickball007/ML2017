# training data

do featurescaling

```python
def featurescaling(X):
    import numpy as np
    X = np.matrix(X)
    normX = np.matrix(X, dtype=float)
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])
    for i in range(X.shape[1]):
        mu[i] = np.mean(X[:, i])
        sigma[i] = np.std(X[:, i])
        normX[:, i] = (X[:, i] - mu[i]) / sigma[i]

    return normX, mu, sigma

normX = featurescaling(X[:, 1:])
X[:, 1:] = normX[0]
X
```

define costfuntion

```python
def costfunction(X, y, theta, londa):
    import numpy as np
    X = np.matrix(X)
    y = np.matrix(y)
    theta = np.matrix(theta)
    m = X.shape[0]

    return (np.sum(np.square((X * theta) - y)) + londa * np.sum(np.square(theta[1:]))) / (2*m)

theta = np.zeros(shape=(163+162, 1))
```

do gradient descent (add adagrad)

```python
def gradientdescent(X, y, theta, alpha, num_iters, londa):
    import numpy as np
    X = np.matrix(X)
    y = np.matrix(y)
    theta = np.matrix(theta)
    m = X.shape[0]
    L_history = np.zeros(num_iters)
    adagrad = np.zeros(theta.shape)
    for i in range(num_iters):
        temp = X.T * (X * theta - y)
        temp[1:] += londa * theta[1:]
        temp = (alpha/m) * temp
        adagrad += np.square(temp)
        temp = temp/np.sqrt(adagrad)
        theta -= temp
        L_history[i] = costfunction(X, y, theta, londa)

    return theta, L_history, adagrad

```

stochastic gradient descent

```python
def SGD(X, y, theta, alpha, num_iters, londa):
    import numpy as np
    X = np.matrix(X)
    y = np.matrix(y)
    theta = np.matrix(theta)
    m = X.shape[0]
    L_history = np.zeros(num_iters)
    adagrad = np.zeros(theta.shape)
    for i in range(num_iters):
        for j in range(m):
            tempX = X[j,:]
            tempy = y[j,:]
            temp = tempX.T * (tempX * theta - tempy)
            temp[1:] += londa * theta[1:]
            temp = (alpha/m) * temp
            adagrad += np.square(temp)
            temp = temp/np.sqrt(adagrad)
            theta -= temp
        L_history[i] = costfunction(X, y, theta, londa)

    return theta, L_history, adagrad
```
